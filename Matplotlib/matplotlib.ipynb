```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[41,52,63,74,55,96,67,48,19]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linestyle='dotted',linewidth=5)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b+--')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle=' ')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle=' ',markersize=20)
```

```python
plt.plot(x,y,'g>',alpha=1)
```
