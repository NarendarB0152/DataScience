```python
from googletrans import Translator
gt_translator=Translator()
en_test=input()
result=gt_translator.translate(en_test,dest='te')
print(result.text)
```

```python
batch=["good morning","This is a test of English to Telugu translation"]
for i in batch:
  print("-",gt_translator.translate(i,dest='te').text)
```

```python
from transformers import pipeline
```

```python
hf_translator=pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi")
hf_result=hf_translator("Hello,how are you today?",max_length=100)
print(hf_result[0]['translation_text'])
```
